```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k,x,y=1,0.0,5.0
def f(x,y,k):
    return -k*y

xx,yy=[],[]
x_max=10
dx=0.1
y0=y
while (x<x_max):
    xx.append(x)
    yy.append(y)
    y=y+dx*f(x,y,k)
    x=x+dx
xx=np.array(xx)
exact=y0*np.exp(-k*xx)

plt.plot(xx,yy,'--',c='red')
plt.plot(xx,exact,c='grey')
```

```python
diff=(exact-yy)
plt.plot(diff,'--',c='black')
```

```python
#initializing variables
e,r,l,x,y=1,0.2,1,0,0

#defining function
def f(x,e,r,l,y):
    return (e-r*y)/l

xx,yy=[],[]
x_max=20
dx=0.1
y0=y

#calculating values
while(x<x_max):
    xx.append(x)
    yy.append(y)
    y=y+dx*f(x,e,r,l,y)
    x=x+dx
    
xx=np.array(xx)
exact=(e/r)*(1-np.exp(-r*xx/l))

plt.plot(xx,yy,'--',c='red')
plt.plot(xx,exact,c='grey')
```

```python
diff=exact-yy
plt.plot(diff,'--',c='black')
```
